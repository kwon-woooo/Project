#2.데이터 가공

```python
#5대범죄에서 필요한 데이터만 불러오기

# 필요없는 열 지우기
for i in range(109,2,-1):
    if (df1.loc[1][i]) !="소계":
        df1 = df1.drop([df1.columns[i]],axis=1)
df1.drop("자치구별(1)", axis=1, inplace = True)

#필요없는 행 지우기
df1.drop([0,1], axis = 0, inplace=True)

# column명 바꿔주기
k = []
for i in range(2014,2023,1):
    for j in ('발생','검거'):
        k.append(str(i)+j)
df1.columns=['자치구']+k
df1.drop(2, inplace = True)

# index 값 자치구 열값으로 바꿔주기
df1=df1.set_index(keys='자치구')

# 수치형으로 만들어 주기
crimes = df1.apply(pd.to_numeric)



#인구수에서 필요한 데이터만 불러오기

# 필요없는 열 지우기
df2 = df2.drop(['동별(1)'],axis=1)
for i in range(108,0,-1):
    if df2.columns[i][-2:] !='.1':
        df2 = df2.drop([df2.columns[i]],axis=1)

#필요없는 행 지우기
df2 = df2.drop([0,1])

# column명 바꿔주기
df2.columns =['자치구',2014,2015,2016,2017,2018,2019,2020,2021,2022]

#index 값 자치구 열값으로 바꿔주기
df2=df2.set_index(keys='자치구')

# 수치형으로 만들어 주기
popul = df2.apply(pd.to_numeric)
```

#4.검거율과 발생률 생성하기(히트맵 시각화까지)

```python
# 검거율 생성

# 빈 데이터프레임 만들기
arrest=pd.DataFrame()

# for문을 이용해 (검거수 / 발생수) 해서 빈 데이터프레임에 열추가하기
for i in range(0,17,2) :
    k=(crimes[crimes.columns[i+1]]) / (crimes[crimes.columns[i]])
    arrest = pd.concat([arrest, k], axis=1)

# column 명 바꾸기
arrest.columns =[2014,2015,2016,2017,2018,2019,2020,2021,2022]

# index값 바꾸기
arrest.index =df1.index
```

```python
# 발생률 생성

# crimes 데이터셋에서 발생 열만 불러오기
crimes_incid = crimes
for i in range(17,0,-1):
    if crimes_incid.columns[i][-2:] !='발생':
        crimes_incid = crimes_incid.drop([crimes_incid.columns[i]],axis=1)

# 발생 열과 인구 데이터 셋 합치기
df22 = pd.concat([crimes_incid,popul],axis=1)

# 수치형 데이터로 바꿔주기
df22=df22.apply(pd.to_numeric)

# 빈 데이터프레임 만들기
incid=pd.DataFrame()

# for문을 이용해 (발생수 / 인구) 해서 빈 데이터프레임에 열추가하기
for i in range(0,9,1) :
    k=(df22[df22.columns[i]]) / (df22[df22.columns[i+9]])
    incid = pd.concat([incid, k], axis=1)

# column 명 바꾸기
incid.columns =[2014,2015,2016,2017,2018,2019,2020,2021,2022]

# index값 바꾸기
incid.index =df1.index
```

#5.발생률과 검거율을 기준으로 년도별 순위매기기 (시각화까지)

```python
# 발생률 점수순위매기기

# 인덱스 순서 알고 있기
incid_s_index=['종로구','중구','용산구','성동구','광진구','동대문구','중랑구','성북구','강북구','도봉구','노원구','은평구','서대문구','마포구','양천구','강서구','구로구','금천구','영등포구','동작구','관악구','서초구','강남구','송파구','강동구']

# 소계 행 제거
incid_s = incid.drop(['소계'],axis=0)

# for문을 이용해 년도 별 순위를 매긴 후 25점부터 1점까지 점수 열 생성 (발생률은 낮은게 좋아서 오름차순 정렬)
for i in incid_s.columns :
    incid_s = incid_s.sort_values(by=[i],axis=0,ascending=True)
    incid_s[str(i)+'년']=range(25,0,-1)

# 원래 순서대로 인덱스 돌리기
incid_s =incid_s.reindex(incid_s_index)

# 년도별 점수 합치기
incid_s_total =incid_s['2014년']+incid_s['2015년']+incid_s['2016년']+incid_s['2017년']+incid_s['2018년']+incid_s['2019년']+incid_s['2020년']+incid_s['2021년']+incid_s['2022년']

# 필요없는 열 삭제
incid_s = incid_s.drop([2014,2015,2016,2017,2018,2019,2020,2021,2022],axis = 1)
```

```python
# 검거율 점수

# 인덱스 순서 알고 있기
arrest_s_index=['종로구','중구','용산구','성동구','광진구','동대문구','중랑구','성북구','강북구','도봉구','노원구','은평구','서대문구','마포구','양천구','강서구','구로구','금천구','영등포구','동작구','관악구','서초구','강남구','송파구','강동구']

# 소계 행 없애기
arrest_s = arrest.drop(['소계'],axis=0)

# for문을 이용해 년도 별 순위를 매긴 후 25점부터 1점까지 점수 열 생성 (검거율은 높은게 좋아서 내림차순 정렬)
for i in arrest_s.columns :
    arrest_s =arrest_s.sort_values(by=[i],axis=0,ascending=False)
    arrest_s[str(i)+'년']=range(25,0,-1)

# 원래 순서대로 인덱스 돌리기
arrest_s = arrest_s.reindex(arrest_s_index)

# 년도별 점수 합치기
arrest_s_total =arrest_s['2014년']+arrest_s['2015년']+arrest_s['2016년']+arrest_s['2017년']+arrest_s['2018년']+arrest_s['2019년']+arrest_s['2020년']+arrest_s['2021년']+arrest_s['2022년']

# 필요없는 열 삭제
arrest_s = arrest_s.drop([2014,2015,2016,2017,2018,2019,2020,2021,2022],axis = 1)
```
